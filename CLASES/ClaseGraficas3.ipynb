```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
año = np.arange(1790,2000,10)
t=np.arange(0,210,10)
real=np.array([3.9,5.3,7.2,9.6,12,17,23,31,38,50,62,75,91,105,122,131,151,179,203,226,249])
Pt= 3.9*np.exp(0.03067*t)
plt.plot(t,real)
plt.plot(t,Pt)
_=plt.xticks(t,año,rotation='vertical')
plt.margins(0.01)
plt.xlabel('t')
plt.ylabel('P(t)')
```

```python
from scipy.integrate import odeint
def dP_dt (P,t):
    dP=2*P
    return dP

t=np.arange(0,3,.1);
x=odeint(dP_dt,10,t)
año = np.arange(1790,2000,10)
#t=np.arange(0,210,10)
real=np.array([3.9,5.3,7.2,9.6,12,17,23,31,38,50,62,75,91,105,122,131,151,179,203,226,249])
Pt= 3.9*np.exp(0.03067*t)
plt.plot(t,real)
plt.plot(t,Pt)
_=plt.xticks(t,año,rotation='vertical')
plt.margins(0.01)
plt.xlabel('t')
plt.ylabel('P(t)')

plt.plot(t,x)
plt.plot(t,real)
plt.plot(t,Pt)
```
